```python
df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
```

```python
df['Ship Mode'].unique()
```

```python
#Transformation 2 - Rename columsn, convert to Lowercase, replace space with underscore
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head(5)
```

```python
#Derive new columns discount, profit, sales price
df['discount'] = df['list_price']*df['discount_percent']*0.01
```

```python
df['sales_price']= df['list_price']-df['discount']
```

```python
df['profit']= df['sales_price']-df['cost_price']
```

```python
df
```

```python
#data type conversion (object to datetime)
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

```python
#Drop a few columns - cost price, list price and discount percent
df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace = True)
```
